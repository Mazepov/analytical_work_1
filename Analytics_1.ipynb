```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches
import warnings

plt.style.use('seaborn')

data = pd.read_csv('courses.csv')
data1 = pd.read_csv('course_contents.csv')
data2 = pd.read_csv('progresses.csv')
data3 = pd.read_csv('progress_phases.csv')
stud = pd.read_csv('students.csv')

#Задаю в переменную today сегодняшнюю дату (для дальнейшей работы с датой)
today = datetime.today()


#ПОДГОТОВКА ДФ

#Переименовываю столбец 'id' в 'course_id'
data.rename(columns={data.columns[1] : 'course_id'}, inplace=True)
#Объединяю два первых датафрема
newdata = pd.merge(data.loc[:, ['course_id', 'title', 'field']], data1, on='course_id')

#Переименовываю столбец 'id' в 'progress_id'
data2.rename(columns={data2.columns[0] : 'progress_id'}, inplace=True)
#Объединяю два вторых датафрема
newdata1 = pd.merge(data2, data3, on='progress_id')
#Объединения всех датафреймов в основной
main_df = pd.merge(newdata1, newdata, on=['course_id', 'module_number', 'lesson_number'])

# Проверка основного дф
# main_df[main_df['student_id'] == '768c2987a744c51ce64a5993a2a94eaf']


#ОСНОВНАЯ ЧАСТЬ
#1.1
#Считаем общее количество курсов в датасете
course_count = len(main_df.groupby('course_id'))

#Считаем количество модулей на каждом курсе
module_count = main_df.groupby('title').max('module_number')
module_count['module_number']

#Считаем количество уроков в каждом модуле на каждом курсе
lessons_count = main_df.loc[:, ['title', 
                                'module_number', 
                                'lesson_number']].groupby(['title', 'module_number']).max('lesson_number')

#Считаем медианное количество уроков в модуле на каждом курсе ( ИСПРАВЛЕНО )
#Считаем на основе lessons_count
lessons_median = lessons_count.reset_index()
lessons_median = lessons_median.loc[:, ['title', 
                                        'lesson_number']].groupby('title').median('lesson_number')
lessons_median.columns = ['median_lessons']

#Считаем количество учеников на каждом курсе
stud_count = pd.DataFrame(main_df.loc[:, ['student_id', 
                                          'title']].drop_duplicates().groupby('title').count()['student_id'])
stud_count.columns = ['student_count']

#Считаем минимальный, максимальный, средний, медианный возраст студентов
stud1 = stud.loc[:, ['id','birthday']].dropna() #Убираем пустые значения
stud1 = stud1[(stud1['birthday'] >= '1940-01-01') & (stud1['birthday'] <= '2004-01-01')] #Фильтруем даты рождения
stud1['Bday'] = pd.to_datetime(stud1['birthday'], format='%Y-%m-%d') #Преобразуем столбец к типу datetime
f = lambda x: today.year - x.year - ((today.month, today.day) < (x.month, x.day)) #Создаем функцию для подсчёта возраста 
                                                                                    #относительно сегодняшней даты
stud1['age'] = stud1['Bday'].apply(f) #Считаем возраст для каждого студента
min_age = stud1['age'].min()
max_age = stud1['age'].max()
mean_age = stud1['age'].mean()
median_age = stud1['age'].median()

#Считаем минимальный, максимальный, средний, медианный возраст студентов на каждом курсе
stud1['student_id'] = stud1['id']

ndata = pd.merge(stud1.loc[:, ['student_id', 
                               'age']], main_df.loc[:, ['student_id', 
                                                        'title']], on='student_id').drop_duplicates()

#Создаю общую таблицу с возрастами ( ИСПРАВЛЕНО )
course_age_min = ndata.groupby('title').min('age')
course_age_min.columns = ['age_min'] #Переименовываю каждый столбец

course_age_max = ndata.groupby('title').max('age')
course_age_max.columns = ['age_max']

course_age_mean = ndata.groupby('title').mean('age')
course_age_mean.columns = ['age_mean']

course_age_median = ndata.groupby('title').median('age')
course_age_median.columns = ['age_median']

course_age_1 = pd.merge(course_age_min, course_age_max, left_index=True, right_index=True)
course_age_2 = pd.merge(course_age_mean, course_age_median, left_index=True, right_index=True)
course_age_total = pd.merge(course_age_1, course_age_2, left_index=True, right_index=True)
```

```python
#1.2 
#Строим bar диаграмму, отражающую количество студентов на каждом курсе
plt.subplots(figsize=(16,8))

plt.bar(stud_count.index, stud_count['student_count'])
plt.title('Количество студентов на каждом курсе', fontsize=14)
plt.xticks(rotation=90, fontsize=14)
plt.yticks(fontsize=12)
plt.show()
```

```python
#1.3
#Строим горизонтальную bar диаграмму ( ИСПРАВЛЕНО )

#Создаём датафрейм с нужными данными
stud_count1 = stud_count.sort_values('student_count', ascending=False)
stud_count1['courses'] = stud_count1.index
field_df = main_df.loc[:, ['title', 'field']].drop_duplicates()
field_df.index = field_df['title']
stud_count_df = pd.merge(field_df['field'], stud_count1, left_index=True, right_index=True)
stud_count_df.sort_values('student_count', ascending=False, inplace=True)

#Строим горизонтальную bar диаграмму, отражающую количество студентов на каждом курсе
plt.subplots(figsize=(16,10))

#Строим диаграмму (alpha = 0.1 не совсем подходит)
sns.barplot(stud_count_df['student_count'], stud_count_df.index, 
            hue=stud_count_df['field'], alpha = 0.7, dodge=False)

#Добавляем название
plt.title('Количество студентов на каждом курсе', fontsize=15, x=0.5, y=1)

#Добавляем линию медианы
plt.vlines(stud_count_df['student_count'].median(), 0, 
           14, color='g', linewidth = 2, linestyle = '--')

# #Убираем рамки
plt.box(on=None)

#Задаём параметры осей и легенды
plt.ylabel(None)
plt.xlabel('Количество студентов')
plt.legend(loc='lower right')
plt.show()

#Игнорирую предупреждение
warnings.filterwarnings('ignore')
```

## 2. Расчет потенциальной нагрузки на преподавателей


```python
#2.1 ( ИСПРАВЛЕНО )
#Расчитываем прирост студентов (создаем необходимый датафрейм для дальнейшего анализа)

# Берём датафрейм с первыми домашними заданиями (Определяем номер модуля и номер урока для каждого курса)
main_df1 = main_df[main_df['is_homework'] == True]
main_df1 = main_df1.loc[:, ['title', 'module_number', 
                            'lesson_number']].drop_duplicates().sort_values(['module_number', 'lesson_number'])
main_df1 = main_df1.drop_duplicates('title', keep='first')

#Создаём новый датасет с первыми номерами модуля и урока для каждого курса
incr = pd.merge(main_df, main_df1, on=('title', 'module_number', 'lesson_number'))

#Далее идёт код без изменений (поменял только main_df на incr)
increase_df = incr.loc[:, ['student_id', 'title', 'start_date']] #Берем необходимые столбцы
increase_df['start_date'] = pd.to_datetime(increase_df['start_date']) #Преобразуем столбец start_date в datetime

#Создаем новый столбец месяца и года (так удобнее в дальнейшей работе)
increase_df['year_month'] = increase_df['start_date'].dt.strftime('%Y-%m')

del increase_df['start_date'] #Удалем ненужный столбец

increase_df1 = increase_df #Создаю копию датафрема

increase_df = increase_df.groupby(['title', 'year_month']).count() #Группируем по месяцам и годам
increase_df_total = increase_df.sort_values('year_month') #Сортируем по дате
```

```python
# 2.2
#Создаем 15 графиков прироста студентов для каждого курса

#Создаем список с title курсов (хотя потом я понял, что их можно было просто взять из датафрейма courses.csv)
title = list(increase_df1['title'].drop_duplicates())

#Создаем общий холс для всех графиков
fig, axs = plt.subplots(15, 1, figsize=(16, 70))

#В цикле делаем графики с учетом всех условий
for i in range(len(title)):
    idf = increase_df1[increase_df1['title'] == title[i]].groupby(['title', 'year_month']).count()
    axs[i] = sns.lineplot(data=idf, x='year_month', y='student_id', hue='title', ci = None, ax = axs[i])
    axs[i].set_xlabel(None)
    axs[i].set_ylabel('Количество студентов')
    axs[i].tick_params(axis='x', labelrotation=35)
    axs[i].legend(loc='upper left')
    
plt.subplots_adjust(hspace=0.3)
plt.show()
```

```python
#2.3
#Строим общий график прироста студентов со всеми курсами

#Создаем общий холс для всех графиков
fig, axs = plt.subplots(figsize=(16, 12))

axs = sns.lineplot(data=increase_df_total, x='year_month', y='student_id', hue='title', ci = None, ax = axs)
plt.title('Изменение количества студентов по годам', fontsize=14)
plt.xticks(rotation=45)
plt.xlabel(None)
plt.ylabel('Количество студентов')

plt.show()
```

```python
#2.4
#Рассчитать количество прогрессов по выполнению домашних работ в каждом месяце
#Фильтруем датафрейм (берем завершенные курсы с дз)
homework_df = main_df[(main_df['is_homework'] == True)] 
homework_df = homework_df.loc[:, ['title', 'start_date', 'finish_date']] #Создаём необходимый для работы датафрейм

#Преобразуем столбец start_date в datetime (секунды приравниваем к нулю)
homework_df['start_date'] = pd.to_datetime(homework_df['start_date'], format='%Y-%m-%d').values.astype('<M8[m]')
#Преобразуем столбец finish_date в datetime
homework_df['finish_date'] = pd.to_datetime(homework_df['finish_date'], format='%Y-%m-%d').values.astype('<M8[m]')

#Заполняем пустые значения в finish_date
homework_df = homework_df.fillna(method='ffill', axis=1)

#Создаю функцию для разделения периода по месяцам
f = lambda x: pd.date_range(x['start_date'], x['finish_date'])
#Создаю вторую функцию для преобразования во множестова с номером месяца и года к каждой строчке
f1 = lambda x: set(x.to_period('m').to_list())

homework_df['aux_col'] = homework_df.apply(f, axis=1) #Разделяю периоды по месяцам
homework_df['month'] = homework_df['aux_col'].apply(f1) #Создаю столбец множеств с месяцами
homework_df = homework_df.loc[:, ['title', 'month']]
homework_df['month'] = homework_df['month'].apply(list)#Преобразую в список
homework_df = homework_df.explode('month') #Создаю новые строчки для каждого месяца

homework_df['month_count'] = 1

#Создаю копию (на всякий случай)
homework_df1 = homework_df
```

```python
#Преобразую солбец month из Period в строку, для дальнейшего построения графиков
homework_df1['month'] = homework_df1['month'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
#Группируем по дате
homework_df_total = homework_df1.groupby(['title', 'month']).count()
# Сортируем по дате
homework_df_total = homework_df_total.sort_values('month')
```

```python
#2.5
#Создаем 15 графиков
#Создаем список с title курсов
title = list(homework_df1['title'].drop_duplicates())

#Создаем общий холст для всех графиков
fig, axs = plt.subplots(15, 1, figsize=(24, 100))

#В цикле делаем графики с учетом всех условий
for i in range(len(title)):
    idf1 = homework_df1[homework_df1['title'] == title[i]].groupby(['title', 'month']).count()
    axs[i] = sns.lineplot(data=idf1, x='month', y='month_count', hue='title', ci = None, ax = axs[i])
    axs[i].set_xlabel(None)
    axs[i].set_ylabel('Количество выполненных работ')
    axs[i].tick_params(axis='x', labelrotation=90)
    axs[i].legend(loc='upper left')
    
plt.subplots_adjust(hspace=0.5)
plt.show()
```

```python
#2.6
#Создаем общий график для всех курсов

#Создаём холст
fig, axs = plt.subplots(figsize=(30, 15))

#Создаём график
axs = sns.lineplot(data=homework_df_total, x='month', y='month_count', hue='title', ci = None, ax = axs)
plt.title('Прогресс по выполненным домашним заданиям', fontsize=14)
plt.xticks(rotation=90)
plt.xlabel(None)
plt.ylabel('Количество выполненных домашних работ')

plt.show()
```

```python
#3.1
#Рассчитать минимальное, максимальное, среднее, медианное время прохождения каждого модуля

#Фильтруем датафрейм (берем завершенные курсы с дз)
module_df = main_df[(main_df['is_homework'] == True) & (main_df['status'] == 'done')] 
#Создаём необходимый для работы датафрейм
module_df = module_df.loc[:, ['title', 'module_title', 'start_date', 'finish_date']]

#Преобразуем столбец start_date в datetime
module_df['start_date'] = pd.to_datetime(module_df['start_date']).dt.to_period('m')
#Преобразуем столбец finish_date в datetime
module_df['finish_date'] = pd.to_datetime(module_df['finish_date']).dt.to_period('m') 

# Считаем сколько месяцов ушло на каждое дз по каждому модулю
module_df['month'] = (module_df['finish_date'].astype(int)- module_df['start_date'].astype(int))+1

module_max = module_df.groupby(['title', 'module_title']).max()
module_min = module_df.groupby(['title', 'module_title']).min()
module_median = module_df.groupby(['title', 'module_title']).median()
module_mean = module_df.groupby(['title', 'module_title']).mean()
```

## 4. Расчет конверсии


```python
#4.1
#Считаем конверсию перехода студентов из одного модуля в другой

#Создаём два дф с окончанием (finish_date) и началом (start_date)
student_finish = main_df.loc[:, ['student_id', 'title', 'module_number', 'lesson_number', 'finish_date', 'status']]
student_start = main_df.loc[:, ['student_id', 'title', 'module_number', 'lesson_number', 'start_date', 'status']]

#Берём только завершенные дз
student_finish = student_finish[student_finish['status'] == 'done']

#Не очень изящно группируем и находим необходимые домашние задания по каждому модулю
student_finish = student_finish.groupby(['title', 'module_number', 'lesson_number']).count()
student_finish = student_finish.reset_index().drop_duplicates(['title', 
                                                               'module_number'], keep='last').loc[:, ['title', 
                                                                                                      'module_number', 
                                                                                                      'student_id']]

student_start = student_start.groupby(['title', 'module_number', 'lesson_number']).count()
student_start = student_start.reset_index().drop_duplicates(['title', 
                                                             'module_number'], keep='last').loc[:, ['title', 
                                                                                                    'module_number', 
                                                                                                    'student_id']]

# #Проверка на основании замечания преподователя
# student_start['finish'] = student_finish['student_id']
# student_start[student_start['title'] == 'Excel Базовый']

#Создаём датафрейм конверсии
stud_conversion = pd.DataFrame()
stud_conversion['title'] = student_start['title']
stud_conversion['module_number'] = student_start['module_number']

#В цикле считаем конверсию для каждого случая (Каждый курс начинается с конверсии равной 1)
list_conv = []
for i in range(student_finish.shape[0]):
  if i == 0:
    list_conv.append(1)   
  elif student_start.iloc[i, 0] != student_start.iloc[i-1, 0]:
    list_conv.append(1)
  else:
    list_conv.append(student_start.iloc[i, 2] / student_finish.iloc[i-1, 2])
stud_conversion['conversion'] = list_conv

# Конверсия по курсам
stud_conversion
```

```python
#4.2

#Строим график конверсии
fig, axs = plt.subplots(figsize=(24, 20))

g = sns.barplot(y = stud_conversion['conversion'], x = stud_conversion['title'], hue=stud_conversion['module_number'])
plt.title('Конверсия перехода студентов из одного модуля в другой', fontsize=15)
plt.xticks(fontsize=12, rotation=90)
plt.xlabel(None)
plt.legend(loc='upper left')
plt.show()
```

```python
#4.3

#Создаем общий холст для всех графиков
fig, axs = plt.subplots(15, 1, figsize=(24, 100))

#В цикле делаем графики с учетом всех условий
for i in range(len(title)):
    idf4 = stud_conversion[stud_conversion['title'] == title[i]]
    axs[i] = sns.barplot(x = idf4['conversion'], y = idf4['title'], hue=idf4['module_number'], ax = axs[i], alpha=0.7)
    axs[i].set_title(title[i])
    axs[i].set_xlabel(None)
    axs[i].set_ylabel(None)
    axs[i].vlines(idf4['conversion'].median(), -0.5, 0.5, color='g', linewidth = 2, linestyle = '--')
    axs[i].axis('off')
```

```python
#4.4
#Выводы по четвертому разделу
print('В четвертом этапе работы мы рассматривали конверсию перехода студентов от одного модуля к другому.\n'
     'Коверсия показана в переменной stud_conversion')
```

Как видно из графиков, в некоторых курсах видны значительное падение, а затем высокий подъем конверсии (четко видно три курса с проблемными модулями). Возможно это говорит о том, что в некоторых случаях студенты пропускаю определенные модули, либо пропускают сдачу домашенего задания. Что говорит о проблемных модулях, соответственно их необходимо переработать.

В целом конверсия на курсах держится на одном уровне. Что является хорошим показателем, и означает что студенты равномерно проходят курс.
На курсе Java разработчик с нуля  видно чёткое уменьшение конверсии от модуля к модулю. Возможно курс является сложным и требует переработки, хотя бы в части домашних заданий.
